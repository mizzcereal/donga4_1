```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

pd.set_option('display.width', 999)           # 출력 너비 무제한
pd.set_option('display.max_columns', 999)     # 모든 열 출력
pd.set_option('display.max_colwidth', 999)

# 1. 데이터셋 로딩 (seaborn에서 제공하는 boston 데이터는 없으므로 외부 링크 사용)
url = "https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv"
df = pd.read_csv(url)
```

```python
df
```

```python
#여러개의 독립변수(방 개수, 평균주택연령, 범죄율를 바탕으로 보스턴 지역으로 주택 가격을 예측하는 다중 선형회구모델구성
#[요구사항1] = 회귀계수를 해석 후 모델의 성능(R**2) 출력

from statsmodels.formula.api import ols
from sklearn.linear_model import LinearRegression

fit = ols('medv ~ crim + zn + indus + chas + nox + rm + age + dis + rad + tax + ptratio + b + lstat', data=df).fit()
```

```python
print("회귀 계수 :")
print(fit.params)
```

```python
print("절편(Intercept) :")
print(fit.params[0])
print("R2(결정계수) :")
print(fit.rsquared)
print("평균제곱오차(MSE) :")
print(mean_squared_error(df.medv, fit.predict(df)))
```
